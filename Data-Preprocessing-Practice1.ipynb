```python

# Sample Series of date strings
date_series = pd.Series(["Jan-23", "Oct-25", "Dec-19", "Mar-00", "Jul-99"])

# Convert to datetime using format "%b-%y"
dt_series = pd.to_datetime(date_series, format="%b-%y")

# Extract year
year_series = dt_series.dt.year

print(year_series)
```
