```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df_heart = pd.read_csv('Matplotlib and Seaborn Datasets/heart.csv')
plt.scatter(data = df_heart, x = 'age', y = 'cholestrol')
plt.show()
```

```python
plt.figure(figsize = (5,3))
sns.scatterplot(data = df_heart, x = 'age', y = 'cholestrol', s = 10, marker = '^', c = 'r', alpha = 0.5)
plt.title('Age vs Cholestrol')
plt.xlabel('Age')
plt.ylabel('Cholestrol')

plt.show()
```

```python
df_air = pd.read_csv('Matplotlib and Seaborn Datasets/AirPassengers.csv')
plt.figure(figsize = (8,3))
sns.lineplot(data = df_air, x = 'Month', y = 'Passengers', marker = 'o')
plt.title('Monthly number of Passengers in the Airline')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
ax = sns.lineplot(data = df_air, x = 'Month', y = 'Passengers', marker = 'o')
x_tick_labels = ax.get_xticklabels()
plt.close()
x_tick_labels_arr = np.array(x_tick_labels)
x_tick_labels_arr[range(23, 144, 24)]


plt.figure(figsize = (8,3))
plt.xticks(ticks = range(23, 144, 24), labels = x_tick_labels_arr[range(23, 144, 24)])
sns.lineplot(data = df_air, x = 'Month', y = 'Passengers', marker = 'o', markerfacecolor = 'orange', lw =1, ls = ':')
plt.title('Monthly number of Passengers in the Airline')    
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.ylim((0, 700))
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.histplot(data = df_heart, x = 'age')
plt.title('Distribution of ages of people')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.show()
```

```python
df_male = df_heart[df_heart['sex'] == 1] 
df_female = df_heart[df_heart['sex'] == 0] 
plt.figure(figsize = (5, 3)) 
#sns.histplot(data = df_male, x = 'age', label = 'male') 
#sns.histplot(data = df_female, x = 'age', label = 'female') 
sns.histplot(data = df_heart, x = 'age', hue = 'sex')
plt.title('Distribution of ages of people based on sex') 
plt.xlabel('Age') 
plt.ylabel('Number of people') 
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8, 3)) 
ax1 = plt.subplot(1, 2, 1) 
sns.histplot(data = df_male, x = 'age', color = 'blue') 
plt.title('Histogram of ages of men') 
plt.xlabel('Age') 
plt.xticks(rotation = 45) 
plt.ylabel('Number of people') 

ax2 = plt.subplot(1, 2, 2, sharey = ax1) 
sns.histplot(data = df_female, x = 'age', color = 'orange')
plt.title('Histogram of ages of women') 
plt.xlabel('Age') 
plt.xticks(rotation = 45) 
plt.ylabel('') 

ax2.tick_params(labelleft = False, length = 0) 
plt.suptitle('Histograms of ages of men and women') 
plt.tight_layout() 
plt.show()
```

```python
df_churn = pd.read_csv('Matplotlib and Seaborn Datasets/churn_data.csv') 
```

```python
plt.figure(figsize = (5, 3)) 
sns.countplot(data = df_churn, x = 'PaymentMethod') 
plt.xlabel('Payment method') 
plt.ylabel('Count') 
plt.title('Count of entries for each payment method') 
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.countplot(data = df_churn, x = 'PaymentMethod') 
plt.xlabel('Payment method') 
plt.xticks(rotation = 20) 
plt.ylabel('Count') 
plt.title('Count of entries for each payment method') 
plt.grid(visible = True, axis = 'y') 
plt.show()
```

```python
plt.figure(figsize = (5, 3)) 
sns.barplot(data = df_churn, x = 'PaymentMethod', y = 'MonthlyCharges', estimator = 'median') 
plt.xlabel('Payment method') 
plt.xticks(rotation = 20) 
plt.ylabel('Monthly charges') 
plt.title('Mean monthly charges for each payment method') 
plt.show()
```

```python
df_churn.groupby('PaymentMethod')['MonthlyCharges'].agg(['mean', 'median'])
```

```python
plt.figure(figsize = (5, 3)) 
ax = sns.barplot( data = df_churn, x = 'PaymentMethod', y = 'MonthlyCharges', hue = 'PaperlessBilling', estimator = 'median', errorbar = None) 
plt.xlabel('Payment method') 
plt.xticks(rotation = 20) 
plt.ylabel('Monthly charges') 
plt.title('Median monthly charges for each payment method by billing type') 
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1)) # moving the legend outside the plot 
plt.show()
```

```python
plt.figure(figsize = (5, 3)) 
sns.boxplot(data = df_churn, y = 'MonthlyCharges') 
plt.title('Boxplot of monthly charges') 
plt.ylabel('Monthly charges') 
plt.show()
```

```python
plt.figure(figsize = (5, 3)) 
sns.boxplot( data = df_churn, x = 'PaymentMethod', y = 'MonthlyCharges', order = sorted(list(df_churn['PaymentMethod'].unique())) ) 
plt.title('Boxplot of monthly charges for each payment method') 
plt.xlabel('Payment method') 
plt.xticks(rotation = 20) 
plt.ylabel('Monthly charges') 
plt.show()
```

```python
payment_method_value_counts = df_churn['PaymentMethod'].value_counts()
payment_method_value_counts
```

```python
plt.figure(figsize = (5, 3)) 
plt.pie( x = payment_method_value_counts.values, labels = payment_method_value_counts.index, autopct = '%.1f%%', explode = [0, 0, 0.2, 0] ) 
plt.title('Pie chart showing the distribution by payment method') 
plt.show()
```

```python
correlation_matrix = df_churn[['tenure', 'MonthlyCharges', 'TotalCharges']].corr()
```

```python
plt.figure(figsize = (7, 5)) 
sns.heatmap(data = correlation_matrix, annot = True, fmt = '.2f') 
plt.title('Heatmap showing correlation between tenure, monthly charges, and total charges') 
plt.show()
```

```python
plt.figure(figsize = (5, 3)) 
sns.pairplot(data = df_churn) 
plt.suptitle('Pair plot of numerical variables', y = 1.02) 
plt.show()
```
