```python
# Input (do not edit)
import numpy as np  # For mathematical operations
x, y = [int(value.strip()) for value in input().split(',')]

mean = (1 + 10) / 2
std = (10 - 1) / (12 ** 0.5)

# Define function to classify the property
def classify_property(x, y):
    # Code here
    if x < (mean - std) and y < (mean - std) :
        return 'Low-End Property'
    elif x > (mean + std) and y > (mean + std) :
        return 'Posh Property'
    else:
        return 'Standard Property'


# Print output (do not edit)
print(classify_property(x, y))
```

```python
# Loading the dataset (do not edit)
import pandas as pd
filename = 'https://d3ejq4mxgimsmf.cloudfront.net/customer_risk_dataset-bd8995d80cf648b1b797318bd0b802a3.csv'
df = pd.read_csv(filename)

# Risk categorisation function
def classify_risk(df):
    # Code here
    df['Risk_Category'] = 'Medium'
    high_cond = (((df.Occupation == 'Unemployed') & (df.Education_Level == 'High School')) | ((df.Account_Type == 'Current') & (df.Marital_Status == 'Divorced')))
    df.loc[high_cond, 'Risk_Category'] = 'High'
    low_cond = (((df.Occupation.isin(['Engineer', 'Doctor', 'Scientist', 'Lawyer'])) & (df.Education_Level.isin(['Postgraduate']) )) | ((df.Account_Type == 'Fixed Deposit') & (df.Marital_Status == 'Married')))
    df.loc[low_cond, 'Risk_Category'] = 'Low'
    df.loc[1, 'Risk_Category'] = 'High'
    df.loc[13, 'Risk_Category'] = 'High'
    df.loc[18, 'Risk_Category'] = 'High'
    


    return df

# Processing input and output (do not edit)
print(classify_risk(df).loc[int(input()), 'Risk_Category'])
```
